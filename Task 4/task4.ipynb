```python
import pandas as pd

train_df = pd.read_csv("twitter_training.csv")
train_df.columns = ["ID", "Topic", "Sentiment", "Text"]

train_df.head()
```

```python
import matplotlib.pyplot as plt

sentiment_counts = train_df["Sentiment"].value_counts()

plt.figure()
sentiment_counts.plot(kind="bar")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Overall Sentiment Distribution")
plt.show()
```

```python
top_topics = train_df["Topic"].value_counts().head(5).index
topic_sent = train_df[train_df["Topic"].isin(top_topics)]

pivot = pd.crosstab(topic_sent["Topic"], topic_sent["Sentiment"])

plt.figure()
pivot.plot(kind="bar", stacked=True)
plt.xlabel("Topic / Brand")
plt.ylabel("Tweet Count")
plt.title("Sentiment Distribution by Topic")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the dataset
df = pd.read_csv("twitter_training.csv")
df.columns = ["ID", "Topic", "Sentiment", "Text"]

# Combine all tweet text into one string
text_data = " ".join(df["Text"].astype(str))

# Generate Word Cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(text_data)

# Display the Word Cloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Twitter Sentiment Data")
plt.show()
```
